```python
#EXPLORE DATA
```

```python
numeric_features=[feature for feature in df_copy.columns if df_copy[feature].dtype !='O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype =='O']

print('We have {} numric:{}'.format(len(numeric_features), numeric_features))
print('We have {} categorical:{}'.format(len(categorical_features), categorical_features))
```

```python
# PROPORTION OF COUNT DATA ON CATOGORICAL COLUMNS
 
for col in categorical_features:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('----------------------------------')
```

```python
# PROPORTION OF COUNT DATA ON NUMERICAL COLUMNS

plt.figure(figsize=(15, 15))
plt.suptitle('UNIVARIATE analysis of numerical features', fontsize=20, fontweight='bold', alpha=0.8,y=1)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],fill=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
# CATOGORICAL COLUMNS
#COUNT OF FREE APPS AND PAID APPS
# COUNT OF PEOPLE(LIKE TEEN,10+,17+,ADULTSAND EVERONE )


plt.figure(figsize=(20, 15))
plt.suptitle('UNIVARIATE analysis of catogorical features', fontsize=20, fontweight='bold', alpha=0.8,y=1)
Category = ['Type', 'Content Rating']

for i in range(0,len(Category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df_copy[Category[i]])
    plt.xlabel(Category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
               
```

```python
# SHOWING PERCENTAGE OF CATEGORY ELEMENT IN PIE CHART 
# IN THIS FAMILY KIND OF APPS ARE VERY MORE IN PLAY STORE 
# WHILE OTHER KINDS OF APPS ARE LESS IN PLAYSTORE

df_copy['Category'].value_counts().plot.pie(y=df_copy['Category'],figsize=(15,16),autopct='%1.1f%%')
plt.show()
```

```python
# TOP 10 APPS CATEGORIES

Category = pd.DataFrame(df_copy['Category'].value_counts())
Category.rename(columns={'Category':'Count'},inplace=True)
Category
```

```python
# which category has largest number of installation?
```

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000     # converting a number into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize=(14,10))
sns.set_context("talk")
sns.set_style('darkgrid')

ax = sns.barplot(x='Category', y= 'Installs', data=df2)
ax.set_xlabel('No. of installation in Billions')
ax.set_ylabel('')
plt.xticks(rotation=45)
ax.set_title('Most popural category in Play Store', size = 20)
```

```python
# WHAT ARE THE TOP 5 MOST INSTALLED APPS IN EACH POPURAL CATEGORIES?
```

```python
dfa = df_copy.groupby(['Category','App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending=False)
apps = ['GAME', 'COMMUNICATION','PRODUCTIVITY','SOCIAL']
sns.set_context('poster')
sns.set_style('darkgrid')

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data=df3, x='Installs', y='App')
    plt.xlabel('Installation in million')
    plt.ylabel('')
    plt.title(app, size= 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

```python
# HOW MANY APPS ARE THERE ON GOOGLE PLAY STORE WHICH GET 5 RATINGS??
```

```python
rating = df_copy.groupby(['Category','Installs','App'])['Rating'].sum().sort_values(ascending=False).reset_index()
top5rating_apps = rating[rating.Rating==5.0]
print('Number of 5 rated apps',top5rating_apps.shape[0])
top5rating_apps.head(5)
```
